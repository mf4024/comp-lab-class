```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
#Read trajectory file
traj = md.load('adp_EM.trr', top='adp_EM.gro')
#compute the phi and psi dihedral angles
phi = md.compute_dihedrals(traj, [[0, 1, 2, 3]])
psi = md.compute_dihedrals(traj, [[1, 2, 3, 4]])

#Make a plot of phi vs time and psi vs time
import mdtraj as md
import matplotlib.pyplot as plt
#Read trajectory file
traj = md.load('adp_exchange4temps.trr', top='adp_exchange4temps.gro')
#compute the phi and psi dihedral angles
phi = md.compute_phi(traj)
psi = md.compute_psi(traj)





#Create log(histogram) of phi vs psi
his, xedges, yedges = np.histogram2d(phi[1][:,0], psi[1][:,0], bins=30, density=True)
free_energy = np.log(his)
#plot free energy using imshow
plt.imshow(free_energy, cmap='Blues', extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]])
plt.xlabel('Phi')
plt.ylabel('Psi')
plt.title('Phi vs Psi')
plt.xlim(-3.14,3.14)
plt.show()
```
