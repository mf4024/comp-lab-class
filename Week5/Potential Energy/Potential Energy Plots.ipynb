```python
import matplotlib.pyplot as plt
import sys
import numpy as np
#Read xvg file
def read_xvg(filename):
    with open(filename) as f:
        lines = f.readlines()
    # remove comments
    lines = [line for line in lines if line[0] not in ('@','#')]
    # split lines
    lines = [line.split() for line in lines]
    # convert to floats
    lines = [[float(x) for x in line] for line in lines]
    # transpose
    lines = np.array(lines).T
    return lines   
read_xvg('energy_T300.xvg')
```

```python
def plot_xvg(filename):
    x,y = read_xvg(filename)
    plt.plot(x,y)
    plt.xlabel('Time (ps)')
    plt.ylabel('Energy)')
    plt.title('Potential Energy(KJ/mol)')
    plt.show()
plot_xvg('energy_T300.xvg')   
#Just for reference 
  
```

```python
def plot_xvg_hist(filename):
    x,y = read_xvg(filename)
    plt.hist(y)
    plt.ylabel('Frequency')
    plt.xlabel('Potential Energy (kJ/mol)')
    plt.title('Potential Energy of protein')
    plt.show()
plot_xvg_hist('energy_T300.xvg')
```

```python
def plot_xvg_hist(filename):
    x,y = read_xvg(filename)
    #change a color
    plt.hist(y, color='purple')
    plt.ylabel('Frequency')
    plt.xlabel('Potential Energy (kJ/mol)')
    plt.title('Potential Energy of protein')
    plt.show()
plot_xvg_hist('energy_T350.xvg')
```

```python
def plot_xvg_hist(filename):
    x,y = read_xvg(filename)
    #change a color
    plt.hist(y, color='green')
    plt.ylabel('Frequency')
    plt.xlabel('Potential Energy (kJ/mol)')
    plt.title('Potential Energy of protein')
    plt.show()
plot_xvg_hist('energy_T400.xvg')
```

```python
def plot_xvg_hist(filename):
    x,y = read_xvg(filename)
    #change a color
    plt.hist(y, color='red')
    plt.ylabel('Frequency')
    plt.xlabel('Potential Energy (kJ/mol)')
    plt.title('Potential Energy of protein')
    plt.show()
plot_xvg_hist('energy_T450.xvg')
```

```python
def plot_xvg_hist_all():
    #change a color
    plt.ylabel('Frequency')
    plt.xlabel('Potential Energy (kJ/mol)')
    plt.title('Potential Energy of protein')
    x,y = read_xvg('energy_T300.xvg')
    plt.hist(y, label='T=300')
    x,y = read_xvg('energy_T350.xvg')
    plt.hist(y, color='purple', label='T=350')
    x,y = read_xvg('energy_T400.xvg')
    plt.hist(y, color='green', label='T=400')
    x,y = read_xvg('energy_T450.xvg')
    plt.hist(y, color='red', label='T=450')
    plt.show()
plot_xvg_hist_all()
```
