```python
import mdtraj
import matplotlib.pyplot as plt

#load and obtain topology from reference structure
r = mdtraj.load('1lys.pdb')
ref_top = r.topology

#select indices for alpha carbons, backbone, and sidechains of chain A
ref_ca = ref_top.select('chainid 0 and name CA')
ref_backbone = ref_top.select('chainid 0 and backbone')
ref_sidechains = ref_top.select('chainid 0 and sidechain')

for i in range(5):
    #load in all of trajectories in one go
    alpha = mdtraj.load(f'relaxed_model_{i+1}_pred_0.pdb')
    alpha_top = alpha.topology
    
    ca = alpha_top.select("name CA")
    backbone = alpha_top.select("backbone")
    sidechains = alpha_top.select("sidechain")

    #alpha carbon rmsd
    print(f'Model {i+1}')
          
    alpha_rmsd = mdtraj.rmsd(alpha, r, frame=0, atom_indices=ca, parallel=True, precentered=False)
    print('RMSD of alpha carbons:', float(alpha_rmsd))

    #backbone rmsd
    backbone_rmsd = mdtraj.rmsd(alpha, r, frame=0, atom_indices=backbone, parallel=True, precentered=False)
    print('RMSD of backbone:', float(backbone_rmsd))

    #sidechain rmsd
    sidechain_rmsd = mdtraj.rmsd(alpha, r, frame=0, atom_indices=sidechains, parallel=True, precentered=False)
    print('RMSD of sidechains:', float(sidechain_rmsd), '\n')
    
    #Plotting the RMSD as bar graph
    rmsd = [float(alpha_rmsd), float(backbone_rmsd), float(sidechain_rmsd)]
    labels = ['Alpha carbons', 'Backbone', 'Sidechains']
    x = range(len(rmsd))
    plt.bar(x, rmsd, tick_label=labels)
    plt.title(f'RMSD of model {i+1}')
    plt.ylabel('RMSD')
    #plt.savefig(f'RMSD of model {i+1}.png')
    plt.show()

```
