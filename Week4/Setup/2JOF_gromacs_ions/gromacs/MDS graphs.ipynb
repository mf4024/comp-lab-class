```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md

# load xtc file
traj = md.load_xtc('ions_center.xtc', top='step5_9.gro')

# function to calculate MSD for a given atom
def msd(traj, atom):
    # chloride
    if atom == 0:
        return md.rmsd(traj, traj, 0)
    # sodium
    elif atom == 1:
        return md.rmsd(traj, traj, 1)
    # oxygen
    elif atom == 2:
        return md.rmsd(traj, traj, 2)

# convert values to integer type
msd_cl = msd(traj, 0).astype(int)
msd_na = msd(traj, 1).astype(int)
msd_o = msd(traj, 2).astype(int)

# plot results vs time
plt.plot(msd_cl, label='Cl')
plt.plot(msd_na, label='Na')
plt.plot(msd_o, label='O')
plt.xlabel('Time (ps)')
plt.ylabel('MSD (nm)')
plt.legend()
plt.title('MSD of Cl, Na, and O vs Time')
plt.show()
```
