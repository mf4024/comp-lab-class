```python
import sys
import numpy as np
import mdtraj as md
import matplotlib.pyplot as plt
import os
#Read xtc file
traj = md.load_xtc('fit_trp-cage.xtc', top='step5_9.gro')

#Plot the end to end distance
plt.plot(md.compute_distances(traj, [[0, 1]]))
plt.xlabel('Time(ns)')
plt.ylabel('Distance (nm)')
plt.title('End to End Distance')
plt.show()



```

```python
#Plot the gyration radius vs time
plt.plot(md.compute_rg(traj))
plt.xlabel('Time(ns)')
plt.ylabel('Radius (nm)')
plt.title('Gyration Radius')
plt.show()
```
