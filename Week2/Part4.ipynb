```python
import matplotlib.pyplot as plt
import sys
import numpy as np
def read_xvg(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()
        data = []
        for line in lines:
            if line.startswith('#'):
                continue
            if line.startswith('@'):
                continue
            if line.startswith('&'):
                continue
            if line.startswith('*'):
                continue
            data.append(line.split())
    return np.array(data, dtype=np.float)
read_xvg("1hz3_T310.run.25000000.energy.xvg")

```

```python
def plot_Potentialxvg(filename):
    data = read_xvg(filename)
    #convert pico second to nanosecond
    data[:,0] = data[:,0] * 0.001
    plt.plot(data[:,0], data[:,1])
    #Plot title
    plt.title("Potential vs Time")
    #X-axis label
    plt.xlabel('Time (ns)')
    #Y-axis label
    plt.ylabel('Potential (J)')
    plt.rcParams["figure.figsize"] = (20,10)
    plt.show()
plot_Potentialxvg("1hz3_T310.run.25000000.energy.xvg")
```

```python
def plot_xvgKinetic(filename):
    data = read_xvg(filename)
    #convert pico second to nanosecond
    data[:,0] = data[:,0] * 0.001
    #convert J/mol to KJ/mol
    data[:,2] = data[:,2] * 0.001
    plt.plot(data[:,0], data[:,2])
    #Plot title
    plt.title("Kinetic En. vs Time")
    #X-axis label
    plt.xlabel('Time (ns)')
    #Y-axis label
    plt.ylabel('Kinetic En. (KJ/mol)')
    plt.show()  
    plt.rcParams["figure.figsize"] = (20,10)    
plot_xvgKinetic("1hz3_T310.run.25000000.energy.xvg")
```

```python
def plot_xvg2and4(filename):
    data = read_xvg(filename)
    #convert pico second to nanosecond
    data[:,0] = data[:,0] * 0.001
    #convert J/mol to KJ/mol
    data[:,2] = data[:,2] * 0.001
    plt.plot(data[:,0], data[:,4], label="Temperature")
    plt.ylim(0, 320)
    #Plot title
    plt.title("Kinetic En. and Temp. vs Time")
    #X-axis label
    plt.xlabel('Time (ns)')
    #Y-axis label
    plt.ylabel('Temp. (K)')
    #Creating an another y-axis
    plt.twinx()
    plt.ylabel('Kinetic En. (KJ/mol)')
    plt.ylim(10, 60)
    #Change a color of the line
    plt.plot(data[:,0], data[:,2], label="Kinetic Energy (KJ/mol)", color="red")
    plt.rcParams["figure.figsize"] = (20,10)
    #Create legend
plot_xvg2and4("1hz3_T310.run.25000000.energy.xvg")
```
