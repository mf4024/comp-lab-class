```python
import numpy as np
import matplotlib.pyplot as plt


#Read in the data from another directory
QMMM = np.loadtxt('../Outputs/monitor/MONITOR-COLVAR.metadynLog')
metad1 = np.loadtxt('../Outputs/metad1_initmonitor/METADYN-COLVAR.metadynLog')
freq=np.loadtxt('../Outputs/metad1_initmonitor_freqhills/METADYN-COLVAR.metadynLog')

#Plot the data
plt.plot(metad1[:,0], metad1[:,1], label='Reaction Coordinate')
plt.title('Reaction Coordinate vs Time_metad')
plt.xlabel('Time (fs)')
plt.ylabel('Reaction Coordinate(d1-d2)')
plt.savefig('RCvsTime_metad1.png')
plt.show()


#Plot the data
plt.plot(QMMM[:,0], QMMM[:,1], label='Reaction Coordinate')
plt.title('Reaction Coordinate vs Time')
plt.xlabel('Time (fs)')
plt.ylabel('Reaction Coordinate(d1-d2)')
plt.savefig('RCvsTime.png')
plt.show()

#Plot the data
plt.plot(freq[:,0], freq[:,1], label='Reaction Coordinate')
plt.title('Reaction Coordinate vs Time_freq')
plt.xlabel('Time (fs)')
plt.ylabel('Reaction Coordinate(d1-d2)')
plt.savefig('RCvsTime_freq.png')
plt.show()
```
