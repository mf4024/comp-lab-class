```python
import numpy as np
import matplotlib.pyplot as plt

#Read in the data from another directory
NPT1 = np.loadtxt('../Outputs/NPT-1.ener')
NVT1 = np.loadtxt('../Outputs/NVT-1.ener')
MONITOR = np.loadtxt('../Outputs/monitor/MONITOR-1.ener')
#Plot the data
plt.plot(NPT1[:,1], NPT1[:,4], label='NPT')
plt.title('Potential Energy vs Time of NPT')
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (Hatree)')
plt.savefig('PEvsTime_NVT.png')
plt.show()

plt.plot(NVT1[:,1], NVT1[:,4], label='NVT')
plt.title('Potential Energy vs Time of NVT')
plt.xlabel('Time (fs)')
plt.ylabel('Potential Energy (Hatree)')
plt.savefig('PEvsTime_NPT.png')
plt.show()
```
