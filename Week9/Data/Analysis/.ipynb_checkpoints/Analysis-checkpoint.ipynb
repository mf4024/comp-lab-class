```python
import MDAnalysis as mda
import MDAnalysis.analysis.rdf as rdf
import MDAnalysis.analysis.msd as msd
import matplotlib.pyplot as plt
import numpy as np



# Load the trajectory
#u1 as temperature = 0.475 u2 as temperature = 0.5 so on
u0475 = mda.Universe("kalj_T0.475_n360_v300_30000000_1.lammpstrj","kalj_T0.475_n360_v300_prod_30000000_1.unwrap.dcd", topology_format="LAMMPSDUMP")
u05 = mda.Universe("kalj_T0.5_n360_v300_10000000_1.lammpstrj","kalj_T0.5_n360_v300_prod_10000000_1.unwrap.dcd", topology_format="LAMMPSDUMP")
u055 = mda.Universe("kalj_T0.55_n360_v300_3000000_1.lammpstrj","kalj_T0.55_n360_v300_prod_3000000_1.unwrap.dcd", topology_format="LAMMPSDUMP")
u06 = mda.Universe("kalj_T0.6_n360_v300_1000000_1.lammpstrj","kalj_T0.6_n360_v300_prod_1000000_1.unwrap.dcd", topology_format="LAMMPSDUMP")
u065 = mda.Universe("kalj_T0.65_n360_v300_350000_1.lammpstrj","kalj_T0.65_n360_v300_prod_350000_1.unwrap.dcd", topology_format="LAMMPSDUMP")
u07 = mda.Universe("kalj_T0.7_n360_v300_180000_1.lammpstrj","kalj_T0.7_n360_v300_prod_180000_1.unwrap.dcd", topology_format="LAMMPSDUMP")
u08 = mda.Universe("kalj_T0.8_n360_v300_150000_1.lammpstrj","kalj_T0.8_n360_v300_prod_150000_1.unwrap.dcd", topology_format="LAMMPSDUMP")
u09 = mda.Universe("kalj_T0.9_n360_v300_150000_1.lammpstrj","kalj_T0.9_n360_v300_prod_150000_1.unwrap.dcd", topology_format="LAMMPSDUMP")
u1 = mda.Universe("kalj_T1_n360_v300_50000_1.lammpstrj","kalj_T1_n360_v300_prod_50000_1.unwrap.dcd", topology_format="LAMMPSDUMP")
u15 = mda.Universe("kalj_T1.5_n360_v300_10000_1.lammpstrj","kalj_T1.5_n360_v300_prod_10000_1.unwrap.dcd", topology_format="LAMMPSDUMP")

# Select the atoms of interest
sel1 = 'type 1'
sel2 = 'type 2'

```

```python
#u1 as temperature = 0.475 u2 as temperature = 0.5 so on
sel = 'type 1'
msd0475 = msd.EinsteinMSD(u0475, sel, msd_type='xyz', fft=True)
msd05 = msd.EinsteinMSD(u05, sel, msd_type='xyz', fft=True)
msd055 = msd.EinsteinMSD(u055, sel, msd_type='xyz', fft=True)
msd06 = msd.EinsteinMSD(u06,sel, msd_type='xyz', fft=True)
msd065 = msd.EinsteinMSD(u065, sel, msd_type='xyz', fft=True)
msd07 = msd.EinsteinMSD(u07, sel, msd_type='xyz', fft=True)
msd08 = msd.EinsteinMSD(u08, sel, msd_type='xyz', fft=True)
msd09 = msd.EinsteinMSD(u09, sel, msd_type='xyz', fft=True)
msd1 = msd.EinsteinMSD(u1, sel, msd_type='xyz', fft=True)
msd15 = msd.EinsteinMSD(u15, sel, msd_type='xyz', fft=True)

msd0475.run()
msd0475_plot = msd0475.results.timeseries
msd05.run()
msd05_plot = msd05.results.timeseries
msd055.run()
msd055_plot = msd055.results.timeseries
msd06.run()
msd06_plot = msd06.results.timeseries
msd065.run()
msd065_plot = msd065.results.timeseries
msd07.run()
msd07_plot = msd07.results.timeseries
msd08.run()
msd08_plot = msd08.results.timeseries
msd09.run()
msd09_plot = msd09.results.timeseries
msd1.run()
msd1_plot = msd1.results.timeseries
msd15.run()
msd15_plot = msd15.results.timeseries

nframes0475 = msd0475.n_frames
timestep = 1 # this needs to be the actual time between frames
lagtimes0475 = np.arange(nframes0475)*timestep # make the lag-time axis

nframes05 = msd05.n_frames
lagtimes05 = np.arange(nframes05)*timestep # make the lag-time axis

nframes055 = msd055.n_frames
lagtimes055 = np.arange(nframes055)*timestep # make the lag-time axis

nframes06 = msd06.n_frames
lagtimes06 = np.arange(nframes06)*timestep # make the lag-time axis

nframes065 = msd065.n_frames
lagtimes065 = np.arange(nframes065)*timestep # make the lag-time axis

nframes07 = msd07.n_frames
lagtimes07 = np.arange(nframes07)*timestep # make the lag-time axis

nframes08 = msd08.n_frames
lagtimes08 = np.arange(nframes08)*timestep # make the lag-time axis

nframes09 = msd09.n_frames
lagtimes09 = np.arange(nframes09)*timestep # make the lag-time axis

nframes1 = msd1.n_frames
lagtimes1 = np.arange(nframes1)*timestep # make the lag-time axis

nframes15 = msd15.n_frames
lagtimes15 = np.arange(nframes15)*timestep # make the lag-time axis


fig = plt.figure()
ax = plt.axes()
#plot the actual MSD
ax.plot(lagtimes0475, msd0475_plot, color='red', linestyle='-', label='T = 0.475')
ax.plot(lagtimes05, msd05_plot, color='blue', linestyle='-', label='T = 0.5')
ax.plot(lagtimes055, msd055_plot, color='green', linestyle='-', label='T = 0.55')
ax.plot(lagtimes06, msd06_plot, color='black', linestyle='-', label='T = 0.6')
ax.plot(lagtimes065, msd065_plot, color='yellow', linestyle='-', label='T = 0.65')
ax.plot(lagtimes07, msd07_plot, color='orange', linestyle='-', label='T = 0.7')
ax.plot(lagtimes08, msd08_plot, color='purple', linestyle='-', label='T = 0.8')
ax.plot(lagtimes09, msd09_plot, color='pink', linestyle='-', label='T = 0.9')
ax.plot(lagtimes1, msd1_plot, color='brown', linestyle='-', label='T = 1')
ax.plot(lagtimes15, msd15_plot, color='grey', linestyle='-', label='T = 1.5')
ax.set(xlabel='lag time (ps)', ylabel='MSD (A^2)',title='MSD of KALJ at different temperatures')
ax.legend()
plt.show()

#I pretty sure I did it wrong. Like very very wrong.
```

```python
#Compute the RDF between the type 1 and type 1 atoms
#r1 as temperature = 0.475 r2 as temperature = 0.5 so on
r0475 = rdf.InterRDF(u0475.select_atoms(sel1), u0475.select_atoms(sel1), nbins=100, range=(0.0, 10.0))
r05 = rdf.InterRDF(u05.select_atoms(sel1), u05.select_atoms(sel1), nbins=100, range=(0.0, 10.0))
r055 = rdf.InterRDF(u055.select_atoms(sel1), u055.select_atoms(sel1), nbins=100, range=(0.0, 10.0))
r06 = rdf.InterRDF(u06.select_atoms(sel1), u06.select_atoms(sel1), nbins=100, range=(0.0, 10.0))
r065 = rdf.InterRDF(u065.select_atoms(sel1), u065.select_atoms(sel1), nbins=100, range=(0.0, 10.0))
r07 = rdf.InterRDF(u07.select_atoms(sel1), u07.select_atoms(sel1), nbins=100, range=(0.0, 10.0))
r08 = rdf.InterRDF(u08.select_atoms(sel1), u08.select_atoms(sel1), nbins=100, range=(0.0, 10.0))
r09 = rdf.InterRDF(u09.select_atoms(sel1), u09.select_atoms(sel1), nbins=100, range=(0.0, 10.0))
r1 = rdf.InterRDF(u1.select_atoms(sel1), u1.select_atoms(sel1), nbins=100, range=(0.0, 10.0))
r15 = rdf.InterRDF(u15.select_atoms(sel1), u15.select_atoms(sel1), nbins=100, range=(0.0, 10.0))

#plot the RDF
r0475.run()
r05.run()
r055.run()
r06.run()
r065.run()
r07.run()
r08.run()
r09.run()
r1.run()
r15.run()

plt.plot(r0475.bins, r0475.rdf, label='T=0.475')
plt.plot(r05.bins, r05.rdf, label='T=0.5')
plt.plot(r055.bins, r055.rdf, label='T=0.55')
plt.plot(r06.bins, r06.rdf, label='T=0.6')
plt.plot(r065.bins, r065.rdf, label='T=0.65')
plt.plot(r07.bins, r07.rdf, label='T=0.7')
plt.plot(r08.bins, r08.rdf, label='T=0.8')
plt.plot(r09.bins, r09.rdf, label='T=0.9')
plt.plot(r1.bins, r1.rdf, label='T=1')
plt.plot(r15.bins, r15.rdf, label='T=1.5')
plt.title('RDF between type 1 and type 1')
plt.xlabel('r (A)')
plt.ylabel('g(r)')
plt.legend()
plt.show()

#Save the Figure
plt.savefig('RDF_type1_type1.png')
```
