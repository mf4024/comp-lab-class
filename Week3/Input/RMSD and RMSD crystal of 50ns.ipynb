```python
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
#Load xvg file
def read_xvg(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()
        data = []
        for line in lines:
            if line.startswith('#'):
                continue
            if line.startswith('@'):
                continue
            if line.startswith('&'):
                continue
            if line.startswith('*'):
                continue
            data.append(line.split())
    return np.array(data, dtype=float)
read_xvg('50ns_rmsd.xvg')
read_xvg('rmsd_50ns_crystal.xvg')
```

```python
data_rmsd = read_xvg('50ns_rmsd.xvg')
plt.plot(data_rmsd[:,0], data_rmsd[:,1])
plt.xlabel('Time (ns)')
plt.ylabel('RMSD (nm)')
plt.title('RMSD')
plt.show()
```

```python
data_rmsd_Crystal=read_xvg('rmsd_50ns_crystal.xvg')
plt.plot(data_rmsd_Crystal[:,0],data_rmsd_Crystal[:,1],color='purple')
plt.xlabel('Time (ns)')
plt.ylabel('RMSD (nm)')
plt.title('RMSD of 50ns with crystal')
plt.show()
# I make two separated plots for 50ns run. Because it will be too noisy for both graphs to be plotted on the same graph.
```
