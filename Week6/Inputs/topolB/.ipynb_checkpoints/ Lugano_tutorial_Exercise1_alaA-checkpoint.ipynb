```python
import sys
import numpy as np
import matplotlib.pyplot as plt

#Read the grid file
grid = np.loadtxt('dialaB.grid.dat')

#Plot the negative of the energy column
plt.scatter(grid[:,0], -grid[:,2])
#label the axes
plt.xlabel('Radian')
plt.ylabel('Energy(KJ/mol)')
plt.title('Energy of ALanine B')
plt.show()
#Save the plot
plt.savefig('Energy_of_dialaB.png')



```
