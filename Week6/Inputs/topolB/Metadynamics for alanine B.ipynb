```python
import sys
import numpy as np
import matplotlib.pyplot as plt

#Read the grid file
grid = np.genfromtxt('fes.dat')

#Plot the negative of the energy column
plt.plot(grid[:,0], grid[:,2])

#label the axes
plt.xlabel('Radian')
plt.ylabel('Energy(KJ/mol)')
plt.title('Estimate of the free energy as a function of the dihedral phi')
plt.show()
#Save the plot
plt.savefig('Free_Energy_AlanineB')
```

```python
grid_1ns = np.genfromtxt('fes_1.dat')
grid_3ns = np.genfromtxt('fes_3.dat')
grid1ns = np.genfromtxt('fes_10.dat')
grid5ns = np.genfromtxt('fes_50.dat')
grid10ns = np.genfromtxt('fes_99.dat')

plt.plot(grid_1ns[:,0], grid_1ns[:,2])
plt.plot(grid_3ns[:,0], grid_3ns[:,2])
plt.plot(grid1ns[:,0], grid1ns[:,2])
plt.plot(grid5ns[:,0], grid5ns[:,2])
plt.plot(grid10ns[:,0], grid10ns[:,2])
plt.xlabel('Radian')
plt.ylabel('Energy(KJ/mol)')
plt.title('Estimates of the free energy as a function of the dihedral phi calculated every 100 Gaussian kernels deposited')
plt.legend(['0.1ns','0.3ns','1ns','5ns','10ns'])
plt.show()
plt.savefig('Free_Energy_AlanineB_all')
```
