```python
#Read in .dat file

import sys
import numpy as np
import matplotlib.pyplot as plt

#Read in file
filename=np.loadtxt('fes.dat')

#Read data before reshaping
phi=filename[:,0]
psi=filename[:,1]
fes=filename[:,2]

#Reshape data
new_phi=phi[:14]
new_psi=psi[::14]
new_fes=fes.reshape(8,14)
plt.title('FES of alanine A')
#plot data with imshow
plt.imshow(new_fes,extent=[new_phi[0],new_phi[-1],new_psi[0],new_psi[-1]],origin='lower')

#Save figure
plt.savefig('fes_alaA.png')
```
