```python
import rdkit
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
import pandas as pd
from rdkit.Chem import PandasTools
from psikit import Psikit
```

```python
%time
file_name = 'smiles.csv'
esol_data = pd.read_csv(file_name)
esol_data.head(1)
#Draw out molecules from smile.csv
smiles = esol_data['smiles']
mols = [Chem.MolFromSmiles(smi) for smi in smiles]

Draw.MolsToGridImage(mols, molsPerRow=2, subImgSize=(200, 200))
```

```python
%time
PandasTools.AddMoleculeColumnToFrame(esol_data, smilesCol='smiles')
esol_data

```

```python
%time
esol_data["n_Atoms"] = esol_data['ROMol'].map(lambda x: x.GetNumAtoms())
esol_data.head(1)
esol_data
```
